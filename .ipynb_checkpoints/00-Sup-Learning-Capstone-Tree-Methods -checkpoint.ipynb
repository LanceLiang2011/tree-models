### Creating Cohorts based on Tenure

**Let's begin by treating each unique tenure length, 1 month, 2 month, 3 month...N months as its own cohort.**

**Treating each unique tenure group as a cohort, calculate the Churn rate (percentage that had Yes Churn) per cohort.**

```python
#CODE HERE
df['Churn_int'] = df.Churn.map({'No':0, 'Yes':1})
```

```python
churn_by_month = df.groupby('tenure').Churn_int.mean()
```

```python
churn_by_month
```

**Now that we have Churn Rate per tenure group 1-72 months, create a plot showing churn rate per months of tenure.**

```python
churn_by_month.plot()
plt.ylabel('Churn Percentage');
```

```python
df.drop(columns='Churn_int', inplace=True)
```

### Broader Cohort Groups
**TASK: Based on the tenure column values, create a new column called Tenure Cohort that creates 4 separate categories:**
   * '0-12 Months'
   * '12-24 Months'
   * '24-48 Months'
   * 'Over 48 Months'    

```python
bins = [0, 12, 24, 48, float('inf')]
labels = ['0-12 Months', '12-24 Months', '24-48 Months', 'Over 48 Months']
df['Tenure_Cohort'] = pd.cut(df.tenure, bins=bins, labels=labels)
```

```python
df[['tenure', 'Tenure_Cohort']].head(10)
```

**Create a scatterplot of Total Charges versus Monthly Charts,colored by Tenure Cohort defined in the previous task.**

```python
#CODE HERE
sns.scatterplot(data=df, x='MonthlyCharges', y='TotalCharges', hue='Tenure_Cohort')
```

**Create a count plot showing the churn count per cohort.**

```python
sns.countplot(data=df,x='Tenure_Cohort', hue='Churn')
```

**Create a grid of Count Plots showing counts per Tenure Cohort, separated out by contract type and colored by the Churn hue.**

```python
sns.catplot(kind='count', data=df, x='Tenure_Cohort', hue='Churn', col='Contract')
```

# Part 4: Predictive Modeling

**Let's explore 4 different tree based methods: A Single Decision Tree, Random Forest, AdaBoost, Gradient Boosting. Feel free to add any other supervised learning models to your comparisons!**


## Single Decision Tree

**Separate out the data into X features and Y label. Create dummy variables where necessary and note which features are not useful and should be dropped.**

```python
y = df.Churn
X = df.drop(columns=['customerID','Churn'])
X = pd.get_dummies(X, drop_first=True)
```

**Perform a train test split, holding out 10% of the data for testing.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

**Decision Tree Perfomance.**
   1. Train a single decision tree model.
   2. Evaluate performance metrics from decision tree, including classification report and plotting a confusion matrix.
   3. Calculate feature importances from the decision tree.
   4. Plot our tree.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import ConfusionMatrixDisplay, classification_report, accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
clf_tree = GridSearchCV(DecisionTreeClassifier(), {'max_depth':[4,5,6,7,8,None]})
```

```python
clf_tree.fit(X_train,y_train)
```

```python
clf_tree.best_params_
```

```python
y_pred = clf_tree.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
feature_importance = clf_tree.best_estimator_.feature_importances_
feature_importance
```

```python
feature_importance_df = pd.DataFrame(feature_importance, index=X.columns, columns=['Importance']).sort_values('Importance', ascending=False)
feature_importance_df
```

```python
feature_importance_df = feature_importance_df[feature_importance_df>0]
```

```python
feature_importance_df.dropna(inplace=True)
```

```python
feature_importance_df
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=feature_importance_df.index, y=feature_importance_df.Importance)
plt.xticks(rotation=90)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8),dpi=150)
plot_tree(clf_tree.best_estimator_,filled=True,feature_names=X.columns);
```

## Random Forest

**TASK: Create a Random Forest model and create a classification report and confusion matrix from its predicted results on the test set.**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf_rf = GridSearchCV(RandomForestClassifier(), {'max_depth': [20,50,70,100,200]})
```

```python
clf_rf.fit(X_train, y_train)
```

```python
clf_rf.best_params_
```

```python
y_pred = clf_rf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```
